```python
import numpy as np
import pandas as pd
import pickle
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = load_breast_cancer()
print(data)
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
df['label'] = data.target
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

```python
df.groupby('label').mean()
```

```python
# Splitting the data into x and y
X = df.drop('label', axis=1)
y = df['label']
```

```python
print(X.shape, y.shape)
```

```python
# Splitting the data into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
row = X_train.iloc[0]
row.to_csv('row_0.csv', index=False)
```

```python
# model training
model = LogisticRegression()

model.fit(X_train, Y_train)

# accuracy
x_train_prediction = model.predict(X_train)

train_acc_score = accuracy_score(Y_train,x_train_prediction)

print('Accuracy on training data : ', train_acc_score)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on test data : ', test_data_accuracy)
```

```python
# Predicting the model
input_data = (16.6,28.08,108.3,858.1,0.08455,0.1023,0.09251,0.05302,0.159,0.05648,0.4564,1.075,3.425,48.55,0.005903,0.03731,0.0473,0.01557,0.01318,0.003892,18.98,34.12,126.7,1124,0.1139,0.3094,0.3403,0.1418,0.2218,0.0782)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one datapoint

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
print(f"input_data_as_numpy_array: {input_data_as_numpy_array}")
print(f"input_data_reshaped: {input_data_reshaped}")
prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
    print('The Breast cancer is Malignant')

else:
    print('The Breast Cancer is Benign')
```
